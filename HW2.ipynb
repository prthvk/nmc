```python
from matplotlib import pyplot as plt
import plotDecBoundaries_2 as pldc
import numpy as np 
import pandas as pd
from scipy.spatial.distance import cdist
```

```python
data = np.genfromtxt("/Users/prithvik/Documents/USC/Assignments/EE559/HW2/HW2_train_set.csv", delimiter=',')
data = data[1:,]
data_points = np.asarray(data[:,:7])
x1 = data[:, 0]
x2 = data[:, 1]
x3 = data[:, 2]
x4 = data[:, 3]
x5 = data[:, 4]
x6 = data[:, 5]
x7 = data[:, 6]
labels = data[:, 7]
sorted_labels = labels[labels.argsort()]
sorted_data = data[data[:, 7].argsort()]
sorted_data_points = np.asarray(sorted_data[:,:7])
```

```python
test_data = np.genfromtxt("/Users/prithvik/Documents/USC/Assignments/EE559/HW2/HW2_test_set.csv", delimiter=',')
test_data = test_data[1:,]
test_data_points = np.asarray(test_data[:,:7])
test_x1 = test_data[:, 0]
test_x2 = test_data[:, 1]
test_x3 = test_data[:, 2]
test_x4 = test_data[:, 3]
test_x5 = test_data[:, 4]
test_x6 = test_data[:, 5]
test_x7 = test_data[:, 6]
test_labels = test_data[:, 7]
test_sorted_labels = test_labels[test_labels.argsort()]
test_sorted_data = test_data[test_data[:, 7].argsort()]
test_sorted_data_points = np.asarray(test_sorted_data[:,:7])
```

```python
sample_means = np.zeros((5,7))
x11, x12, x13, x14, x15, x16, x17, c1 = 0,0,0,0,0,0,0,0
x21, x22, x23, x24, x25, x26, x27, c2 = 0,0,0,0,0,0,0,0
x31, x32, x33, x34, x35, x36, x37, c3 = 0,0,0,0,0,0,0,0
x41, x42, x43, x44, x45, x46, x47, c4 = 0,0,0,0,0,0,0,0
x51, x52, x53, x54, x55, x56, x57, c5 = 0,0,0,0,0,0,0,0

for i in range(len(data)):
    if data[i][7] == 0:
        x11 = x11+data[i][0]
        x12 = x12+data[i][1]
        x13 = x13+data[i][2]
        x14 = x14+data[i][3]
        x15 = x15+data[i][4]
        x16 = x16+data[i][5]
        x17 = x17+data[i][6]
        c1 = c1 + 1
    elif data[i][7] == 1:
        x21 = x21+data[i][0]
        x22 = x22+data[i][1]
        x23 = x23+data[i][2]
        x24 = x24+data[i][3]
        x25 = x25+data[i][4]
        x26 = x26+data[i][5]
        x27 = x27+data[i][6]
        c2 = c2 + 1
    elif data[i][7] == 2:
        x31 = x31+data[i][0]
        x32 = x32+data[i][1]
        x33 = x33+data[i][2]
        x34 = x34+data[i][3]
        x35 = x35+data[i][4]
        x36 = x36+data[i][5]
        x37 = x37+data[i][6]
        c3 = c3 + 1
    elif data[i][7] == 3:
        x41 = x41+data[i][0]
        x42 = x42+data[i][1]
        x43 = x43+data[i][2]
        x44 = x44+data[i][3]
        x45 = x45+data[i][4]
        x46 = x46+data[i][5]
        x47 = x47+data[i][6]
        c4 = c4 + 1
    elif data[i][7] == 4:
        x51 = x51+data[i][0]
        x52 = x52+data[i][1]
        x53 = x53+data[i][2]
        x54 = x54+data[i][3]
        x55 = x55+data[i][4]
        x56 = x56+data[i][5]
        x57 = x57+data[i][6]
        c5 = c5 + 1

sample_means[0,0], sample_means[0,1], sample_means[0,2], sample_means[0,3], sample_means[0,4], sample_means[0,5], sample_means[0,6], = x11/c1, x12/c1, x13/c1, x14/c1, x15/c1, x16/c1, x17/c1
sample_means[1,0], sample_means[1,1], sample_means[1,2], sample_means[1,3], sample_means[1,4], sample_means[1,5], sample_means[1,6], = x21/c2, x22/c2, x23/c2, x24/c2, x25/c2, x26/c2, x27/c2
sample_means[2,0], sample_means[2,1], sample_means[2,2], sample_means[2,3], sample_means[2,4], sample_means[2,5], sample_means[2,6], = x31/c3, x32/c3, x33/c3, x34/c3, x35/c3, x36/c3, x37/c3
sample_means[3,0], sample_means[3,1], sample_means[3,2], sample_means[3,3], sample_means[3,4], sample_means[3,5], sample_means[3,6], = x41/c4, x42/c4, x43/c4, x44/c4, x45/c4, x46/c4, x47/c4
sample_means[4,0], sample_means[4,1], sample_means[4,2], sample_means[4,3], sample_means[4,4], sample_means[4,5], sample_means[4,6], = x51/c5, x52/c5, x53/c5, x54/c5, x55/c5, x56/c5, x57/c5
```

```python
dist_mat = cdist(data_points, sample_means)
pred_label = np.argmin(dist_mat, axis=1)
pred_label[:] = pred_label[:]

err_count = 0
for i in range(len(labels)):
    if (int(pred_label[i])!=int(labels[i])):
        err_count = err_count+1
acc_rate = (1-(err_count/len(data_points)))*100
print("The Classification Accuraacy for Training Data using all 7 features = " + str(acc_rate) + "%")

test_dist_mat = cdist(test_data_points, sample_means)
test_pred_label = np.argmin(test_dist_mat, axis=1)
test_pred_label[:] = test_pred_label[:]

test_err_count = 0
for i in range(len(test_labels)):
    if (int(test_pred_label[i])!=int(test_labels[i])):
        test_err_count = test_err_count+1
test_acc_rate = (1-(test_err_count/len(test_data_points)))*100
print("The Classification Accuraacy for Testing Data using all 7 features = " + str(test_acc_rate) + "%")
```
